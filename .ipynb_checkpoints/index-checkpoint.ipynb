```python
# código obligatorio --> cabecera
from IPython.display import IFrame
```

```python
# asignamos al sitio web un nombre, este será llamado link
link = "https://es.wikipedia.org/wiki/WWE_Championship" 
IFrame(link, width=900, height=900)
```

```python
# importamos "panda" --> permite usar graficos y diversas funciones
import pandas as pd

tablaInicial = pd.read_html(link, # recordar que link --> url de la página web (wikipedia)
                        header=0, # where is the header?
                        flavor='bs4')
```

```python
# comando que nos ayuda a contar cuántass tablas tenemos en el link insertado
len(tablaInicial)
```

```python
# tablaInicial[3] --> significa el cuadro N°4, ya que recordamos que en phyton el contador empieza en 0
copiaTablaInicial = tablaInicial[3].copy()
```

```python
# invocamos a la copia de "tablaInicial", precisamente la tabla N°4
copiaTablaInicial
```

```python
# función que nos dice los tipos de variables que tiene cada columna
copiaTablaInicial.info()
```

```python
# Paso 1: Recordamos que "copiaTablaInicial" es el nombre de la tabla seleccionada
# Paso 2: .iloc --> función a la cuál le indicaremos cuáles son las COLUMNAS con las que nos queremos quedar
# Paso 3: le decimos a la función cuales son las columnas con las que deseamos trabajar [1,2,4,5,6]
copiaTablaInicial = copiaTablaInicial.iloc[:, [1, 2, 4, 5, 6]]
```

```python
# invocamos esta nueva tabla
copiaTablaInicial

# como podemos observar, la columna con título "Evento" y "Notas y referencias" fueron eliminados
```

```python
# utilizamos la función ".rename" para cambiarle el nombre a las columnas
copiaTablaInicial = copiaTablaInicial.rename(columns = {'Año': 'Year', 
                                                        'Lucha': 'Fecha', 
                                                        'Lucha.2': 'Ubicacion', 
                                                        'Estadísticas': 'Reinado',
                                                        'Estadísticas.1': 'Días'})
```

```python
# invocamos a la tabla actualizada
copiaTablaInicial
```

```python
# como podemos observar, la fila N°1 contiene información repetida, por tanto la eliminaremos
# (salió así porque en wikipedia era un cuadro de doble entrada y se mezcló en el jupyter notebook)

#si usamos el siguiente código, los índices no se reinician
#copiaTablaInicial = copiaTablaInicial.drop(0, axis=0)

# con este código, los índices sí se reinician --> importante usar ".reset_index(drop=True)"
copiaTablaInicial = copiaTablaInicial.drop(0, axis=0).reset_index(drop=True)
```

```python
# invocamos a la tabla actualizada (sin la fila N°0 que se repetía con el nombre de las columnas)
copiaTablaInicial
```

```python
# nos aseguramos de que las celdas con texto no tengan espacios al final ni al principio

# para ello usamos ".strip()" --> Función que elimina los espacios en blanco, ya que estos cuentan como carácteres

# el método str.strip() en Pandas se puede usar para eliminar los espacios en blanco al inicio y al final de cada valor en una columna determinada
copiaTablaInicial['Campeones'] = copiaTablaInicial['Campeones'].str.strip()
copiaTablaInicial['Fecha'] = copiaTablaInicial['Fecha'].str.strip()
copiaTablaInicial['Ubicacion'] = copiaTablaInicial['Ubicacion'].str.strip()
copiaTablaInicial['Reinado'] = copiaTablaInicial['Reinado'].str.strip()
copiaTablaInicial['Días'] = copiaTablaInicial['Días'].str.strip()
```

```python
# invocamos a la tabla actualizada, la cual si es que existían espacios en blanco en los datos de las celdas, se eliminarían
copiaTablaInicial
```

```python
# detectamos la presencia de símbolos en los datos numéricos que no sean numéricos o puntuales

# usamos la función ".str.contains" para detectar símbolos más allá de [^] alfanuméricos (\w) o espacios (\s)
# es decir, esta función nos sirve para poder detectar caracteres diferentes a los espacios y números (^, :, \, ~)
copiaTablaInicial.Campeones[copiaTablaInicial.Campeones.str.contains(pat=r'[^\w\s]', regex=True)]

# detectamos si hay elementos no númeroricos en la celda "Campeones"
```

```python
# detectamos si hay elementos no númeroricos en la celda "Fecha"
copiaTablaInicial.Fecha[copiaTablaInicial.Fecha.str.contains(pat=r'[^\w\s]', regex=True)]
```

```python
# detectamos si hay elementos no númeroricos en la celda "Ubicación"
copiaTablaInicial.Ubicacion[copiaTablaInicial.Ubicacion.str.contains(pat=r'[^\w\s]', regex=True)]
```

```python
# detectamos si hay elementos no númeroricos en la celda "Reinado"
copiaTablaInicial.Reinado[copiaTablaInicial.Reinado.str.contains(pat=r'[^\w\s]', regex=True)]
```

```python
# detectamos si hay elementos no númeroricos en la celda "Días"
copiaTablaInicial.Días[copiaTablaInicial.Días.str.contains(pat=r'[^\w\s]', regex=True)]
```

```python
# nos aseguramos de que no haya espacios como parte de los nombres de las columnas
columnas_con_espacios = copiaTablaInicial.columns[copiaTablaInicial.columns.str.contains(' ')]
```

```python
# invocamos a "columnas_con_espacios" y observamos si es que los nombres de las columnas tienen espacios en blanco
columnas_con_espacios

# lo que nos sale dentro de los corchetes ['variable'] --> indica en que variables encontramos aquello que le pedimos
# en el código anterior, lo cual le pedimos que busque espacios en blanco
```

```python
# nos deshacemos de cualquier valor detectado en el paso anterior
copiaTablaInicial.columns = copiaTablaInicial.columns.str.replace(' ', '')

# le pedimos que remplace los espacios en blanco por el vacío (no espacios)
```

```python
# creamos una copia de la tabla que hasta ahora tenemos con las 5 columnas (Campeones, Fecha, Ubicación, Reinado, Días)
copiaTablaInicial_NuevaCopia = copiaTablaInicial.copy()
```

```python
# mantenemos solo el valor de los días en la columna "Días"
copiaTablaInicial_NuevaCopia['Días'] = copiaTablaInicial_NuevaCopia.Días.str.replace(pat=r'\D', repl='', regex=True)
```

```python
# vemos la nueva tabla
copiaTablaInicial_NuevaCopia
```

```python
# nos aseguramos que se hayan borrado los carácteres no numéricos existentes (+, -) --> "Días"
copiaTablaInicial_NuevaCopia.Días[copiaTablaInicial_NuevaCopia.Días.str.contains(pat=r'[^\w\s]', regex=True)]

# como observamos, quítamos los caracteres no numéricos
```

```python
# mantenemos solo el valor de los días en la columna "Reinado"
copiaTablaInicial_NuevaCopia['Reinado'] = copiaTablaInicial_NuevaCopia.Reinado.str.replace(pat=r'\D', repl='', regex=True)
```

```python
# nos aseguramos que se hayan borrado los carácteres no numéricos existentes (+, -) --> "Reinado"
copiaTablaInicial_NuevaCopia.Reinado[copiaTablaInicial_NuevaCopia.Reinado.str.contains(pat=r'[^\w\s]', regex=True)]

# como observamos, quítamos los caracteres no numéricos
```

```python
# vemos la nueva tabla
copiaTablaInicial_NuevaCopia
```

```python
# nos aseguramos que se hayan borrado los carácteres no numéricos existentes (+, -) --> "Campeones"
copiaTablaInicial_NuevaCopia['Campeones'] = copiaTablaInicial_NuevaCopia['Campeones'].str.replace('.', '', regex=False)

# como observamos, usando la función ".str.replace("caracterQueNoQueremos", "CaracterRemplazar")"
```

```python
# vemos la nueva tabla
copiaTablaInicial_NuevaCopia
```

```python
# nos aseguramos que se hayan borrado los carácteres no numéricos existentes (.) --> "Campeones"
copiaTablaInicial_NuevaCopia.Campeones[copiaTablaInicial_NuevaCopia.Campeones.str.contains(pat=r'[^\w\s]', regex=True)]

# como observamos, quítamos los caracteres no numéricos como "."
```

```python
# función para obtener las palabras antes de la coma en la columna "Columna"
copiaTablaInicial_NuevaCopia['Ubicacion'] = copiaTablaInicial_NuevaCopia['Ubicacion'].apply(lambda x: x.split(',')[0].strip())
```

```python
# nos aseguramos que se hayan borrado los carácteres presentes después de la primera "," detectada
copiaTablaInicial_NuevaCopia

# como observamos, quítamos los caracteres después ","
```

```python
# detectamos si hay elementos no númeroricos en la celda "Ubicación"
copiaTablaInicial_NuevaCopia.Ubicacion[copiaTablaInicial_NuevaCopia.Ubicacion.str.contains(pat=r'[^\w\s]', regex=True)]
```

```python
# eliminamos los "."
copiaTablaInicial_NuevaCopia['Ubicacion'] = copiaTablaInicial_NuevaCopia['Ubicacion'].str.replace('.', '', regex=False)
```

```python
# eliminamos los "-"
copiaTablaInicial_NuevaCopia['Ubicacion'] = copiaTablaInicial_NuevaCopia['Ubicacion'].str.replace('—', '', regex=False)
```

```python
# vemos si es que se borraron los "." y "-"
copiaTablaInicial_NuevaCopia.Ubicacion[copiaTablaInicial_NuevaCopia.Ubicacion.str.contains(pat=r'[^\w\s]', regex=True)]

# como observamos, eliminamos todos los caractéres no numéricos (. y —)
```

```python
# observamos como va quedando la tabla
copiaTablaInicial_NuevaCopia
```
